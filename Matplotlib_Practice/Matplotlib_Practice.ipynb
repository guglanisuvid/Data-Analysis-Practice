```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4], linestyle = '--', color = 'red', marker = 's', linewidth = '2')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Line Plot', fontdict={'fontname': 'calibri', 'fontweight': 20, 'color': 'green', 'fontsize': 20})

plt.savefig('line plot.png')

plt.show()
```

```python
plt.figure()
plt.plot([2, 4, 5, 6, 7, 10])
plt.show()

plt.figure()
plt.plot([1, 3, 6, 5, 8, 19])
plt.show()
```

```python
plt.figure()

plt.suptitle('Main title of the plot')

plt.subplot(3, 1, 1)
plt.plot([1, 2, 4, 5, 7, 8, 10, 11])
plt.title('Row 1')
plt.xlabel('Row 1 xlabel')
plt.ylabel('Row 1 ylabel')

plt.subplot(3, 1, 2)
plt.plot([5, 6, 8, 9, 11, 12, 14, 15])
plt.title('Row 2')
plt.xlabel('Row 2 xlabel')
plt.ylabel('Row 2 ylabel')

plt.subplot(3, 1, 3)
plt.plot([10, 11, 13, 14, 16, 17, 19, 20])
plt.title('Row 3')
plt.xlabel('Row 3 xlabel')
plt.ylabel('Row 3 ylabel')

plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(3, 1)

plt.suptitle('Main title of the plot')

axes[0].plot([1, 5, 3, 7, 5, 9, 7])
axes[0].set_title('Row 1')
axes[0].set_xlabel('Row 1 xlabel')
axes[0].set_ylabel('Row 1 ylabel')

axes[1].plot([7, 9, 8, 10, 9, 11, 10])
axes[1].set_title('Row 2')
axes[1].set_xlabel('Row 2 xlabel')
axes[1].set_ylabel('Row 2 ylabel')

axes[2].plot([1, 3, 4, 5, 6, 8, 0])
axes[2].set_title('Row 3')
axes[2].set_xlabel('Row 3 xlabel')
axes[2].set_ylabel('Row 3 ylabel')

plt.tight_layout()

plt.show()
```

```python
ax = plt.axes(projection = '3d')

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 4, 5, 6, 7, 8, 9, 10]
z = [5, 6, 7, 8, 9, 10, 11, 12]

ax.plot(x, y, z)

plt.show()
```

```python
ax = plt.axes(projection = '3d')

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 4, 5, 6, 7, 8, 9, 10]
z = [5, 6, 7, 8, 9, 10, 11, 12]
dx = np.ones(8)
dy = np.ones(8)
dz = np.random.randint(1, 10, 8)

ax.bar3d(x, y, z, dx, dy, dz)

plt.show()
```

```python
ax = plt.axes(projection = '3d')

X = [1, 2, 3, 4, 5, 6, 7, 8]
Y = [3, 4, 5, 6, 7, 8, 9, 10]
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z)

plt.show()
```
