# Data Cleaning

```python
df = pd.read_excel(r"C:\Users\manan\Documents\Pandas_Practice\Datasets\Customer Call List.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['Not_Useful_Column'], inplace = True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('N')
df['Paying Customer'] = df['Paying Customer'].replace({'Yes': 'Y', 'No': 'N', 'N/a': 'N'})
```

```python
df.drop_duplicates(inplace = True)
```

```python
df["First_Name"] = df["First_Name"].str.replace(r"[^a-zA-Z]", "", regex=True)
df["Last_Name"] = df["Last_Name"].str.replace(r"[^a-zA-Z]", "", regex=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].replace(r'[^0-9]', '', regex = True)
df.dropna(subset=['Phone_Number'], inplace = True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df = df[df['Phone_Number'].apply(lambda x: len(x) == 12)]
```

```python
df.drop(columns = ['CustomerID', 'Address'], inplace = True)
```

```python
df = df[df['Do_Not_Contact'].apply(lambda x: str(x) == 'N')]
```

```python
df.sort_values(by = ['First_Name'], inplace = True)
```

```python
df.reset_index(drop = True, inplace = True)
df
```
